```python
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline
```

```python
from tensorflow import layers
```

```python
num_iterations = 1000
learning_rate = 0.001
```

```python
df = pd.read_csv('Data/WitWorkers-Dataset.csv')
```

```python
df.head()
```

```python
df = df.drop(df.columns[0],1)
```

```python
df.head()
```

```python
df = df.fillna(df.mean())
```

```python
#data

features = np.array(df[df.columns[1:-1]])
target = np.array(df[df.columns[-1]]).reshape([-1,1])
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
features_scaled = scaler.fit_transform(features)
```

```python
#placeholders

X = tf.placeholder(tf.float32, [None, 7])
y = tf.placeholder(tf.float32, [None, 1])
```

```python

layer1 = tf.layers.dense(tf.layers.flatten(X), units=30, activation=tf.nn.relu)
#layer = tf.nn.sigmoid(layer)
layer2 = tf.layers.dense(layer1, units=50, activation=tf.nn.relu)
#layer = tf.nn.sigmoid(layer)
output = tf.layers.dense(layer2, 1)
```

```python
optimizer = tf.train.AdamOptimizer(learning_rate=learning_rate)
```

```python
loss = tf.losses.mean_squared_error(y, output)
train = optimizer.minimize(loss)
```

```python
saver = tf.train.Saver()
```

```python
with tf.Session() as sess:
    sess.run(tf.global_variables_initializer())
    
    for it in range(60000):
        _, mse= sess.run([train, loss], feed_dict={X:features_scaled, y:target})
        print('Interation: {} MSE: {}'.format(it, mse))
        
    saver.save(sess,'./model.ckpt')
```

```python
with tf.Session() as sess:
    saver.restore(sess, './model.ckpt')
    out = sess.run(output, feed_dict={X:features[-4:]})
    
```
